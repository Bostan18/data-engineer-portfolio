# 📊 Analyse des livres scrapés

Objectif : visualiser la distribution des prix, notes et disponibilité du catalogue.

```python
import sqlite3
import matplotlib.pyplot as plt
import pandas as pd

conn = sqlite3.connect('../data/books.db')
df = pd.read_sql_query('SELECT * FROM books', conn)
conn.close()

print(f'{len(df)} livres chargés')
df.head()
```

## 1. Distribution des prix

```python
plt.figure(figsize=(10, 5))
plt.hist(df['price'], bins=20, color='steelblue', edgecolor='white')
plt.axvline(df['price'].mean(), color='red', linestyle='--', label=f'Moyenne: £{df["price"].mean():.2f}')
plt.axvline(df['price'].median(), color='green', linestyle='--', label=f'Médiane: £{df["price"].median():.2f}')
plt.xlabel('Prix (£)')
plt.ylabel('Nombre de livres')
plt.title('Distribution des prix')
plt.legend()
plt.tight_layout()
plt.savefig('../data/prix_distribution.png', dpi=100)
plt.show()

print(f'Prix min: £{df["price"].min():.2f}')
print(f'Prix max: £{df["price"].max():.2f}')
print(f'Prix moyen: £{df["price"].mean():.2f}')
```

## 2. Répartition des notes

```python
ratings = df['rating'].value_counts().sort_index()

plt.figure(figsize=(8, 5))
bars = plt.bar(ratings.index.astype(str) + '★', ratings.values, color=['gray', 'tomato', 'orange', 'gold', 'limegreen'])
for bar, val in zip(bars, ratings.values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.3, str(val), ha='center')
plt.xlabel('Note')
plt.ylabel('Nombre de livres')
plt.title('Répartition des notes')
plt.tight_layout()
plt.savefig('../data/ratings_distribution.png', dpi=100)
plt.show()
```

## 3. Disponibilité (en stock / rupture)

```python
stock = df['in_stock'].value_counts()
en_stock = stock.get(1, 0)
out_stock = stock.get(0, 0)
labels = ['En stock', 'Rupture']
values = [en_stock, out_stock]

plt.figure(figsize=(6, 6))
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90, colors=['limegreen', 'tomato'])
plt.title('Disponibilité des livres')
plt.tight_layout()
plt.savefig('../data/stock_pie.png', dpi=100)
plt.show()

print(f'En stock: {en_stock} | Rupture: {out_stock}')
```

## 4. Prix moyen par note

```python
mean_price_by_rating = df.groupby('rating')['price'].mean()

plt.figure(figsize=(8, 5))
plt.bar(mean_price_by_rating.index.astype(str) + '★', mean_price_by_rating.values, color=['gray', 'tomato', 'orange', 'gold', 'limegreen'])
plt.xlabel('Note')
plt.ylabel('Prix moyen (£)')
plt.title('Prix moyen par note')
plt.tight_layout()
plt.savefig('../data/prix_par_note.png', dpi=100)
plt.show()

for rating, price in mean_price_by_rating.items():
    print(f'{int(rating)}★: £{price:.2f}')
```
